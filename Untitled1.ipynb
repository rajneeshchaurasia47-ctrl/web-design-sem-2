```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go

# --- Helper Function for Graphing (Question 5 & 6) ---
def make_graph(stock_data, title):
    fig = go.Figure()

    # Plot Closing Price
    fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data['Close'], mode='lines', name='Close Price'))

    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis_title='Stock Price (USD)',
        height=600,
        width=1000
    )

    fig.show()

# --- Question 1: Tesla Stock Data (yfinance) ---
print("Fetching Tesla Stock Data...")
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset index, save, and display first 5 rows
tesla_data = tesla_data.reset_index()
print("Tesla Data (First 5 rows):")
display(tesla_data.head())
# NOTE: Screenshot isliye lena hai ki yeh output dikhe

# --- Question 2: Tesla Revenue Data (Webscraping) ---
print("\nFetching Tesla Revenue Data...")
tesla_url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
tesla_html = requests.get(tesla_url).text
tesla_soup = BeautifulSoup(tesla_html, "html5lib")

tesla_table = tesla_soup.find("table", {"id": "q1"}) # Macrotrends usually has specific table IDs, trying to find revenue table
# Note: Sometimes tables change structure. Macrotrends has a table with class "table" or specific ID.
# Let's try a robust search for the revenue table
tesla_revenue_table = None
for table in tesla_soup.find_all("table"):
    if "Revenue" in table.text or "q1" in str(table.get('id', '')):
        tesla_revenue_table = table
        break

if tesla_revenue_table:
    tesla_revenue = pd.read_html(str(tesla_revenue_table))[0]
    # Clean up if needed (often the first row is header)
    if 'Date' not in tesla_revenue.columns:
        # Try to parse if structure is different
        tesla_revenue = pd.read_html(str(tesla_revenue_table))[0]

    print("Tesla Revenue (Last 5 rows):")
    display(tesla_revenue.tail())
else:
    print("Could not scrape Tesla revenue table automatically. Check the URL structure.")
    tesla_revenue = pd.DataFrame() # Placeholder

# --- Question 3: GameStop Stock Data (yfinance) ---
print("\nFetching GameStop Stock Data...")
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Reset index, save, and display first 5 rows
gme_data = gme_data.reset_index()
print("GameStop Data (First 5 rows):")
display(gme_data.head())
# NOTE: Screenshot isliye lena hai ki yeh output dikhe

# --- Question 4: GME Revenue Data (Webscraping) ---
print("\nFetching GameStop Revenue Data...")
gme_url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
gme_html = requests.get(gme_url).text
gme_soup = BeautifulSoup(gme_html, "html5lib")

gme_revenue_table = None
for table in gme_soup.find_all("table"):
    if "Revenue" in table.text or "q1" in str(table.get('id', '')):
        gme_revenue_table = table
        break

if gme_revenue_table:
    gme_revenue = pd.read_html(str(gme_revenue_table))[0]
    print("GameStop Revenue (Last 5 rows):")
    display(gme_revenue.tail())
else:
    print("Could not scrape GME revenue table automatically.")
    gme_revenue = pd.DataFrame()

# --- Question 5: Plot Tesla Stock Graph ---
print("\nGenerating Tesla Graph...")
make_graph(tesla_data, "Tesla (TSLA) Stock Price History")
# NOTE: Graph ke screenshot ko upload karna hai

# --- Question 6: Plot GameStop Stock Graph ---
print("\nGenerating GameStop Graph...")
make_graph(gme_data, "GameStop (GME) Stock Price History")
# NOTE: Graph ke screenshot ko upload karna hai

print("\nAll steps completed successfully!")
```
